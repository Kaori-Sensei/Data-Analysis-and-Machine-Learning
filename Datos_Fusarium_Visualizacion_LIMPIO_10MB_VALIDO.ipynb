## Preparar los datos geográficos

### Subtask:
Convertir las columnas de "Latitude" y "Longitude" del DataFrame a un formato GeoDataFrame para trabajar con datos espaciales.


**Reasoning**:
Import the geopandas library, drop rows with missing latitude or longitude, create a GeoDataFrame from the cleaned DataFrame using the 'Longitude' and 'Latitude' columns, and display the head of the GeoDataFrame.



```python
import geopandas as gpd
from shapely.geometry import Point

df_mich_cleaned = df_mich.dropna(subset=['Latitude', 'Longitude'])

geometry = [Point(xy) for xy in zip(df_mich_cleaned['Longitude'], df_mich_cleaned['Latitude'])]
gdf_mich = gpd.GeoDataFrame(df_mich_cleaned, geometry=geometry, crs='EPSG:4326')

display(gdf_mich.head())
```

## Realizar kernel density estimation

### Subtask:
Calcular la densidad de puntos o la intensidad de las variables ('Incidencia', 'Severidad') utilizando Kernel Density Estimation. Esto generará una superficie ráster o contornos.


**Reasoning**:
Import necessary libraries, extract coordinates and the variable of interest ('Incidencia'), and perform Kernel Density Estimation using `scipy.stats.gaussian_kde`. Define a grid for the output and evaluate the KDE on this grid.



```python
import numpy as np
from scipy.stats import gaussian_kde

# Ensure gdf_muestras is defined and has 'Incidencia'
if 'gdf_muestras' in locals() or 'gdf_muestras' in globals():
    if gdf_muestras is not None and not gdf_muestras.empty and 'Incidencia' in gdf_muestras.columns:

        # Extract coordinates (Longitude and Latitude)
        coords = np.vstack([gdf_muestras.geometry.x, gdf_muestras.geometry.y])

        # Extract the variable values ('Incidencia')
        incidencia_values = gdf_muestras['Incidencia'].values

        # Perform Kernel Density Estimation weighted by 'Incidencia'
        # We use the coordinates and the values to estimate the intensity
        try:
            kde = gaussian_kde(coords, weights=incidencia_values)

            # Define a grid for the output surface
            # Get the bounds of the points with some padding
            x_min, y_min, x_max, y_max = gdf_muestras.total_bounds
            x_range = x_max - x_min
            y_range = y_max - y_min

            # Define grid resolution (adjust as needed for desired detail vs performance)
            grid_points = 100 # Number of points in each dimension

            x_grid = np.linspace(x_min - x_range * 0.1, x_max + x_range * 0.1, grid_points)
            y_grid = np.linspace(y_min - y_range * 0.1, y_max + y_range * 0.1, grid_points)
            X, Y = np.meshgrid(x_grid, y_grid)
            xy = np.vstack([X.ravel(), Y.ravel()])

            # Evaluate the KDE on the grid
            Z = kde(xy).reshape(X.shape)

            print("Kernel Density Estimation for Incidencia calculated successfully.")
            # Store results for visualization
            kde_incidencia_grid = Z
            kde_incidencia_extent = [x_grid[0], x_grid[-1], y_grid[0], y_grid[-1]]

        except Exception as e:
            print(f"An error occurred during Kernel Density Estimation: {e}")
            kde_incidencia_grid = None
            kde_incidencia_extent = None

    else:
        print("Error: gdf_muestras is not available, empty, or missing 'Incidencia' column. Cannot perform KDE.")
        kde_incidencia_grid = None
        kde_incidencia_extent = None

else:
    print("Error: gdf_muestras GeoDataFrame is not defined. Please run the previous steps to load your Excel data and convert it to a GeoDataFrame.")
    kde_incidencia_grid = None
    kde_incidencia_extent = None
```

## Preparar y exportar datos a KML

### Subtask:
Asegurar que ambos GeoDataFrames (`michoacan_municipalities` y `gdf_muestras`) estén en el mismo sistema de coordenadas (EPSG:4326) y exportarlos a un archivo KML.

**Reasoning**:
Ensure both GeoDataFrames are in EPSG:4326 CRS for KML compatibility, and then export them to a single KML file. KML supports multiple layers, so exporting each GeoDataFrame to the same KML file will create separate layers for municipalities and points.

```python
import geopandas as gpd
import os

# Ensure michoacan_municipalities and gdf_muestras are defined and not empty
if ('michoacan_municipalities' in locals() or 'michoacan_municipalities' in globals()) and \
   ('gdf_muestras' in locals() or 'gdf_muestras' in globals()):

    if michoacan_municipalities is not None and not michoacan_municipalities.empty and \
       gdf_muestras is not None and not gdf_muestras.empty:

        # Ensure both GeoDataFrames are in WGS84 (EPSG:4326) CRS, which is standard for KML
        if michoacan_municipalities.crs != "EPSG:4326":
            michoacan_municipalities = michoacan_municipalities.to_crs("EPSG:4326")
            print("michoacan_municipalities CRS converted to EPSG:4326.")

        if gdf_muestras.crs != "EPSG:4326":
            gdf_muestras = gdf_muestras.to_crs("EPSG:4326")
            print("gdf_muestras CRS converted to EPSG:4326.")


        # Define the output KML file path
        output_kml_file = "michoacan_municipios_puntos.kml"

        # Export both GeoDataFrames to the same KML file.
        # geopandas can write multiple layers to a KML file.
        try:
            # Export municipalities layer using pyogrio engine
            michoacan_municipalities.to_file(output_kml_file, driver='KML', layer='Municipios_Michoacan', engine='pyogrio')

            # Export points layer, appending to the existing KML file using pyogrio engine
            gdf_muestras.to_file(output_kml_file, driver='KML', layer='Puntos_Muestreo', mode='a', engine='pyogrio')

            print(f"Archivo KML '{output_kml_file}' generado correctamente con capas de municipios y puntos.")

            # Store the output file name for download step
            kml_combined_file = output_kml_file

        except Exception as e:
            print(f"Ocurrió un error al exportar a KML: {e}")
            kml_combined_file = None # Indicate export failed

    else:
        print("Error: One or both GeoDataFrames (michoacan_municipalities, gdf_muestras) are empty or None. Cannot export to KML.")
        kml_combined_file = None # Indicate export failed

else:
    print("Error: One or both required GeoDataFrames (michoacan_municipalities, gdf_muestras) are not defined. Please run the previous steps.")
    kml_combined_file = None # Indicate export failed
```

## Añadir capa de puntos de muestreo con Severidad (Tamaño)

### Subtask:
Agregar los puntos de muestreo al mapa de folium utilizando círculos cuyo tamaño varíe según la 'Severidad' y con popups informativos.

**Reasoning**:
Iterate through the `gdf_muestras` GeoDataFrame and add a CircleMarker for each point to the folium map, scaling the radius by 'Severidad'. Include information about 'MUESTRA', 'Incidencia', and 'Severidad' in a popup for each marker.

```python
import folium
from folium.plugins import HeatMap

# Ensure m (folium map object) and gdf_muestras are defined

if 'm' in locals() and ('gdf_muestras' in locals() or 'gdf_muestras' in globals()):
    if gdf_muestras is not None and not gdf_muestras.empty:

        # Add a layer for the sample points with size based on Severidad
        severity_points_layer = folium.FeatureGroup(name='Puntos de Muestreo (Severidad)').add_to(m)

        # Determine a scaling factor for the circle radius based on 'Severidad'
        # Adjust this factor as needed to make the circles visually appropriate
        max_severidad = gdf_muestras['Severidad'].max()
        # Avoid division by zero if all severidad values are 0
        severity_scale_factor = 10 if max_severidad == 0 else 50 / max_severidad # Scale max severity to a radius of 50 (adjust 50 as needed)

        # Iterate through each point in the gdf_muestras GeoDataFrame
        for index, row in gdf_muestras.iterrows():
            # Create a popup with relevant information
            popup_html = f"""
            <b>Muestra:</b> {row['MUESTRA']}<br>
            <b>Incidencia:</b> {row['Incidencia']}<br>
            <b>Severidad:</b> {row['Severidad']}<br>
            <b>Latitud:</b> {row['Latitude']}<br>
            <b>Longitud:</b> {row['Longitude']}
            """
            # Determine circle radius based on Severidad, adding a small base size if needed for visibility of 0 values
            radius = (row['Severidad'] + 1) * severity_scale_factor if row['Severidad'] == 0 else row['Severidad'] * severity_scale_factor
            # Ensure a minimum radius for visibility
            radius = max(radius, 5) # Minimum radius of 5 (adjust as needed)


            # Add a CircleMarker for each point, with radius based on Severidad
            folium.CircleMarker(
                location=[row['Latitude'], row['Longitude']],
                radius=radius, # Set radius based on scaled Severidad
                color='red', # Outline color
                fill=True,
                fill_color='red', # Fill color
                fill_opacity=0.6, # Transparency
                popup=folium.Popup(popup_html, max_width=300), # Add popup with HTML content
            ).add_to(severity_points_layer)

        print("Capa de puntos de muestreo (Severidad) añadida al mapa.")

        # The Heatmap for Incidencia will be added in the next step
        # The Layer Control will be added after all layers are included

        # Display the map with the added municipal and points layers
        display(m)

    else:
        print("Error: gdf_muestras GeoDataFrame is empty or None. Cannot add sample point layer.")
else:
    print("Error: El objeto mapa 'm' o el GeoDataFrame 'gdf_muestras' no están definidos. Por favor, asegúrate de ejecutar los pasos anteriores.")
```

## Cómo interpretar el mapa interactivo de Fusarium en Michoacán

Este mapa interactivo muestra la distribución espacial de los puntos de muestreo en el estado de Michoacán, visualizando la Incidencia y Severidad de Fusarium utilizando diferentes elementos:

---

### Interpretación de la Severidad (Tamaño del Punto)

La **Severidad** de Fusarium en cada punto de muestreo se representa por el **tamaño del círculo** que marca la ubicación del punto.

*   **Círculos más grandes** indican una **mayor severidad**.
*   **Círculos más pequeños** indican una **menor severidad**.

Para obtener el valor exacto de Severidad en un punto específico, puedes hacer clic en el círculo correspondiente en el mapa interactivo. Aparecerá un cuadro de información (popup) con los detalles de esa muestra, incluyendo el valor de Severidad.

---

### Interpretación de la Incidencia (Mapa de Calor)

La **Incidencia** de Fusarium se visualiza como una **capa de mapa de calor** superpuesta en el mapa. Esta capa estimada te da una idea general de las áreas donde la incidencia es más alta o más baja.

*   Las áreas con **colores más intensos o brillantes** en el mapa de calor (por defecto, suelen ir hacia el rojo o amarillo brillante) indican una **mayor concentración o "densidad" estimada de la incidencia** en esa zona, basándose en los puntos de muestreo cercanos.
*   Las áreas con **colores más suaves, tenues o transparentes** indican una **menor incidencia estimada** en esa área.

El mapa de calor suaviza los valores de los puntos individuales para mostrar patrones espaciales generales de la incidencia. Al igual que con la Severidad, puedes hacer clic en los círculos de los puntos de muestreo para ver el valor de Incidencia exacto registrado en ese lugar.

---

Utiliza el **Control de Capas** (generalmente un icono en una de las esquinas del mapa) para activar o desactivar la visualización de los "Puntos de Muestreo (Severidad)" y el "Mapa de Calor de Incidencia".
